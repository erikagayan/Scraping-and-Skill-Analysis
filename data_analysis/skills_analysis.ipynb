```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("skills_count.csv")
df.head()
```

```python
df_sorted = df.sort_values('Count', ascending=False)
df_sorted.head()
```

```python
plt.figure(figsize=(15, 10))
plt.bar(df_sorted['Skill'], df_sorted['Count'])
plt.title('Skills count')
plt.xlabel('Skills')
plt.ylabel('Number')
plt.xticks(rotation=90)
plt.show()
```
